```python
df_dropped = df

upper_limit = df_dropped['UNIT_PRICE'].quantile(0.999)
df_dropped = df_dropped[(df_dropped['UNIT_PRICE'] < upper_limit) & (df_dropped['UNIT_PRICE'] > 0)] # drop outliers
#plt.hist(df['UNIT_PRICE'][(df['UNIT_PRICE'] < upper_limit) & (df['UNIT_PRICE'] > 0)], bins=100)

upper_limit = df_dropped['DELIVERED_PRICE'].quantile(0.999)
df_dropped = df_dropped[(df_dropped['DELIVERED_PRICE'] < upper_limit) & (df_dropped['DELIVERED_PRICE'] > 0)]
#plt.hist(df['DELIVERED_PRICE'][(df['DELIVERED_PRICE'] < 300) & (df['DELIVERED_PRICE'] > 0)], bins=100)

upper_limit = df_dropped['DELIVERED_QUANTITY'].quantile(0.999)
df_dropped = df_dropped[(df_dropped['DELIVERED_QUANTITY'] < upper_limit) & (df_dropped['DELIVERED_QUANTITY'] > 0)]
#plt.hist(df['DELIVERED_QUANTITY'][(df['DELIVERED_QUANTITY'] < upper_limit) & (df['DELIVERED_QUANTITY'] > 0)], bins=100)
```

```python
dropped_vals = df_dropped
dropped_vals['NORM_UNI_PRICE'] = (2 * ((dropped_vals['UNIT_PRICE'] - min(dropped_vals['UNIT_PRICE'])) / (max(dropped_vals['UNIT_PRICE']) - min(dropped_vals['UNIT_PRICE'])))) - 1

dropped_vals['NORM_DEL_PRICE'] = (2 * ((dropped_vals['DELIVERED_PRICE'] - min(dropped_vals['DELIVERED_PRICE'])) / (max(dropped_vals['DELIVERED_PRICE']) - min(dropped_vals['DELIVERED_PRICE'])))) - 1

dropped_vals['NORM_DEL_QUANT'] = (2 * ((dropped_vals['DELIVERED_QUANTITY'] - min(dropped_vals['DELIVERED_QUANTITY'])) / (max(dropped_vals['DELIVERED_QUANTITY']) - min(dropped_vals['DELIVERED_QUANTITY'])))) - 1
```

```python
cleaned = dropped_vals[['NORM_UNI_PRICE', 'NORM_DEL_PRICE', 'NORM_DEL_QUANT']]
cleaned.describe()
```

```python
x_train = cleaned[0:400000]
x_test = cleaned[400000:]

#x_train.shape[1:]
```

```python
from tensorflow.keras.layers import Input, Flatten, Dense, Reshape, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.regularizers import l2

class Autoencoder:
  def __init__(self, input_dim, inner_activation): # 128 64 2 64 128
    self.inner_dim = 2
    self.input_dim = input_dim
    self.encoder_layers = []
    self.encoder_layers.append(Dense(64,activation='relu',name='encoder1'))
    self.encoder_layers.append(Dense(16,activation='relu',name='encoder2'))
    self.encoder_layers.append(Dense(self.inner_dim,activation=inner_activation,name='encoder3') ) #sparse would have regularization on here

    self.decoder_layers = []
    self.decoder_layers.append(Dense(16,activation='relu',name='decoder1'))
    self.decoder_layers.append(Dense(64,activation='relu',name='decoder2'))
    x = input_dim[0]
    for i in range(1,len(input_dim)):
      x = x * input_dim[i]
    self.decoder_layers.append(Dense(x,activation=None,name='decoder3'))
  
  def get_autoencoder(self):
    """ Builds the full autoencoder model with encoder and decoder. """
    inputs = Input(self.input_dim,name='autoencoder_input')
    x = Flatten()(inputs) #no need to flatten?
    for i in range(len(self.encoder_layers)):
      x = self.encoder_layers[i](x)
    for i in range(len(self.decoder_layers)):
      x = self.decoder_layers[i](x)
    outputs = Reshape(self.input_dim)(x)
    return Model(inputs=inputs,outputs=outputs)
  
  def get_encoder(self):
    """ Builds just the encoder model. """
    inputs = Input(self.input_dim,name='encoder_input')
    x = Flatten()(inputs) #no need to flatten?
    for i in range(len(self.encoder_layers)):
      x = self.encoder_layers[i](x)
    return Model(inputs=inputs,outputs=x)
  
  def get_decoder(self):
    """ Builds just the decoder model. """
    embedding = Input((self.inner_dim,),name='decoder_input')
    x = embedding
    for i in range(len(self.decoder_layers)):
      x = self.decoder_layers[i](x)
    outputs = Reshape(self.input_dim)(x)
    return Model(inputs=embedding,outputs=outputs)

autoencoder = Autoencoder((3,) , 'tanh') # input_dim? (3,) ? , inner_activation

ae_model = autoencoder.get_autoencoder()
encoder_model = autoencoder.get_encoder()
decoder_model = autoencoder.get_decoder()

ae_model.compile(SGD(0.01,momentum=0.9),loss='mean_squared_error') # mse is better
print(ae_model.summary())
```

```python
history = ae_model.fit(x_train,x_train,batch_size=32,epochs=10,verbose=False,validation_split=0.1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
